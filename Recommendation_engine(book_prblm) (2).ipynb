```python
df.columns=['userId','book','Rating']
```

```python
df
```

```python
df.drop_duplicates(keep=False,inplace=True)#dropping duplicate rows
```

```python
df.reset_index(drop = True)
```

```python
df.info()
```

```python
user_book_df = df.pivot_table(index='userId',
                                 columns='book',
                                 values='Rating').reset_index(drop=True)
```

```python
user_book_df
```

```python
len(user_book_df.columns)
```

```python
user_book_df.index = df.userId.unique()
```

```python
user_book_df
```

```python
#Impute those NaNs with 0 values
user_book_df.fillna(0, inplace=True)
```

```python
user_book_df.head(50)
```

```python
user_book_df.isnull().sum()
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
